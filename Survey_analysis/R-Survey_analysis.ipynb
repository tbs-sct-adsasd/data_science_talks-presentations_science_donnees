```r
library(corrplot)
library(zeallot)
library(psych)
library(factoextra)
library(ltm)
options(repr.plot.width=10, repr.plot.height=10)
df<-read.table("PCA.csv", header=TRUE, sep=",")
df
```

```r
#check data quality ( no missing values, scale seems correct etc. )
summary(df)
```

```r
#missing value check
#is.na(df)
```

Check scatterplots and correlation matrix for linear relationships 

```r
#scatterplot matrices
pairs(df)
```

```r
#correlation table to verify linear relationships
M<-cor(df, method="pearson")
corrplot(M, method= "number")
```

Use Bartlett and KMO to verify suitability of data for PCA. 
Bartlett's test result is statistically significant.
KMO overall is 0.83, individual >0.72 . 

```r
bartlett.test(df,df)
```

```r
KMO(df)
```

All assumptions have been met, continue to PCA

```r
df.pca<- prcomp(df, center=TRUE, scale. =TRUE)
summary(df.pca)
```

We can now try to narrow down the PCA selection. 
Eigenvalue one criterion suggests 5 components, scree plot and interpretability suggest 4. Four components retained

```r
#components with eigenvalues less than 1 should be eliminated
eigen.val<-get_eigenvalue(df.pca)
eigen.val
```

```r
#inflection point at component 5, take components appearing before inflection
fviz_eig(df.pca)
```

```r
#loading table 

var_coord_func <- function(loadings, comp.sdev){
  loadings*comp.sdev
    #source http://www.sthda.com/english/articles/31-principal-component-methods-in-r-practical-guide/118-principal-component-analysis-in-r-prcomp-vs-princomp/
}

loadings<-df.pca$rotation
sdev<- df.pca$sdev
varcoord<- t(apply(loadings,1,var_coord_func, sdev))
format(round(varcoord[,1:5],2), nsmall=2)
```
